Obtener los trazadores cúbicos para los siguientes puntos y evaluar el valor en 0,5

Captura de pantalla 2024-11-06 152553.png

la funcion base para verificacion es f(x) = 0,1x^4 - 0,2x^3 - x^2 + x +1

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

# Definir los puntos base (x, f(x)) de la tabla
x_puntos = np.array([0, 1, 2, 3, 4])
y_puntos = np.array([1, 0.9, -1, -2.3, 1.8])

# Crear el trazador cúbico usando CubicSpline
spline_cubico = CubicSpline(x_puntos, y_puntos)

# Evaluar el valor en x = 0.5
x_val = 0.5
f_0_5 = spline_cubico(x_val)

# Mostrar el resultado
print(f"El valor de f(0.5) usando el trazador cúbico es: {f_0_5}")

# Graficar los puntos y el trazador cúbico
x_grafica = np.linspace(min(x_puntos), max(x_puntos), 100)
y_grafica = spline_cubico(x_grafica)

plt.plot(x_puntos, y_puntos, 'o', label="Puntos dados")
plt.plot(x_grafica, y_grafica, label="Trazador cúbico")
plt.plot(x_val, f_0_5, 'ro', label=f"f(0.5) = {f_0_5:.2f}")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.title("Interpolación por trazadores cúbicos")
plt.show()
```
