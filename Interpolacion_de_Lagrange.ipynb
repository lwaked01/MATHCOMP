Por medio de polinomios de interpolación de Lagrange realice la estimación del valor de f(2,5), para los grados 1,2 y 3, utilizando como puntos bases:

Tambien halle los polinomios de grado 1 y 2 a los puntos seleccionados

```python
import sympy as sp

# Definir los puntos base (x, f(x)) de la tabla
puntos = [(0, 0.5), (1, 0.1), (2, 1), (3, 2.1), (4, 2.4)]

# Variable simbólica
x = sp.symbols('x')

def polinomio_lagrange(puntos, grado):
    """Calcula el polinomio de Lagrange de un cierto grado con los puntos dados."""
    n = len(puntos)
    if grado >= n:
        raise ValueError("El grado no puede ser mayor o igual al número de puntos.")

    # Seleccionar los puntos necesarios para el grado dado
    puntos_seleccionados = puntos[:grado + 1]
    polinomio = 0

    # Calcular el polinomio de Lagrange
    for i in range(grado + 1):
        xi, yi = puntos_seleccionados[i]
        termino = yi
        for j in range(grado + 1):
            if i != j:
                xj, _ = puntos_seleccionados[j]
                termino *= (x - xj) / (xi - xj)
        polinomio += termino

    return sp.simplify(polinomio)

# Calcular los polinomios de grado 1, 2 y 3
polinomio_grado_1 = polinomio_lagrange(puntos, 1)
polinomio_grado_2 = polinomio_lagrange(puntos, 2)
polinomio_grado_3 = polinomio_lagrange(puntos, 3)

# Mostrar los polinomios
print("Polinomio de grado 1:", polinomio_grado_1)
print("Polinomio de grado 2:", polinomio_grado_2)
print("Polinomio de grado 3:", polinomio_grado_3)

# Evaluar f(2.5) en cada polinomio
x_val = 2.5
f_2_5_grado_1 = polinomio_grado_1.subs(x, x_val)
f_2_5_grado_2 = polinomio_grado_2.subs(x, x_val)
f_2_5_grado_3 = polinomio_grado_3.subs(x, x_val)

# Mostrar los resultados
print(f"\nEstimación de f(2.5) con el polinomio de grado 1: {f_2_5_grado_1}")
print(f"Estimación de f(2.5) con el polinomio de grado 2: {f_2_5_grado_2}")
print(f"Estimación de f(2.5) con el polinomio de grado 3: {f_2_5_grado_3}")
```
