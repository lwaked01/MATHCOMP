Cree un programa que ajuste a un polinomio de segundo grado los datos dados en la tabla. El progranma debe calcular ademas el coeficiente de correlación


```python
import numpy as np
from scipy.stats import pearsonr
import matplotlib.pyplot as plt

# Datos de la tabla
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([-0.9, 0, 2, 4.5, 8.3, 13, 18])

# Ajuste del polinomio de segundo grado
coeficientes = np.polyfit(x, y, 2)
polinomio = np.poly1d(coeficientes)

# Generar valores ajustados
y_ajustado = polinomio(x)

# Imprimir los coeficientes del polinomio
print("Coeficientes del polinomio de segundo grado:", coeficientes)

# Calcular el coeficiente de correlación entre x y y
corr_x_y, _ = pearsonr(y, y_ajustado)
print("\nCoeficiente de correlación (r):", corr_x_y)

# Graficar los datos y el polinomio ajustado
plt.scatter(x, y, color="blue", label="Datos originales")
plt.plot(x, y_ajustado, color="red", label="Ajuste polinómico")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Ajuste de polinomio de segundo grado")
plt.show()
```
