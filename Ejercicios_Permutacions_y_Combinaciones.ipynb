```python


def ordenar_libros(matematicas, fisica, quimica):
    return math.factorial(matematicas + fisica + quimica) // (math.factorial(matematicas) * math.factorial(fisica) * math.factorial(quimica))

print("6a. Formas de ordenar libros (todos juntos):", ordenar_libros(8, 6, 2))

def ordenar_libros_matematicas_juntos(matematicas, fisica, quimica):
    return math.factorial(matematicas) * math.factorial(fisica + quimica + 1) // (math.factorial(fisica) * math.factorial(quimica) * math.factorial(1))

print("6b. Formas de ordenar libros (matemáticas juntos):", ordenar_libros_matematicas_juntos(8, 6, 2))
```

# Ejercicio 11

```python
def comite_techno_magico(techno, magico, t_seleccionados, m_seleccionados):
    total = 0
    for t in range(t_seleccionados + 1):
        for m in range(m_seleccionados + 1):
            if t + m == 5:
                total += math.comb(techno, t) * math.comb(magico, m)
    return total

print("11a. Comité de 2 technoarquitectos y 3 arquitectos mágicos:", comite_techno_magico(5, 7, 2, 3))

def comite_techno_magico_b(techno, magico, t_seleccionados, m_seleccionados):
    return math.comb(techno, t_seleccionados) * math.comb(magico - 1, m_seleccionados - 1)

print("11b. Comité con un arquitecto mágico determinado:", comite_techno_magico_b(5, 7, 2, 3))

def comite_techno_magico_c(techno, magico, t_seleccionados, m_seleccionados):
    return math.comb(techno - 2, t_seleccionados) * math.comb(magico, m_seleccionados)

print("11c. Comité sin dos technoarquitectos determinados:", comite_techno_magico_c(5, 7, 2, 3))
```

# Ejercicio 20

```python
def evaluar_expresiones():
    expresiones = [
        ("9!=9*8*7*6!", lambda: math.factorial(9) == 9*8*7*math.factorial(6)),
        ("5!*4!=20!", lambda: math.factorial(5)*math.factorial(4) == math.factorial(20)),
        ("5!+5!=10!", lambda: math.factorial(5)+math.factorial(5) == math.factorial(10)),
        ("8!=9!/9", lambda: math.factorial(8) == math.factorial(9)/9),
        ("5!=10!/2!", lambda: math.factorial(5) == math.factorial(10)/math.factorial(2)),
        ("3/4+1/4=1!", lambda: 3/4 + 1/4 == math.factorial(1)),
        ("3!*2!=6!", lambda: math.factorial(3)*math.factorial(2) == math.factorial(6)),
        ("8!=6!*56", lambda: math.factorial(8) == math.factorial(6)*56)
    ]
    resultados = []
    for expr, func in expresiones:
        try:
            resultado = func()
            resultados.append(f"{expr}: {'Verdadero' if resultado else 'Falso'}")
        except:
            resultados.append(f"{expr}: Error")
    return resultados

print("20. Evaluación de expresiones:")
for resultado in evaluar_expresiones():
    print(resultado)
```

```python
def seleccionar_focos_defectuosos(total, defectuosos, seleccionados):
    a = math.comb(total - defectuosos, seleccionados)
    b = math.comb(defectuosos, 1) * math.comb(total - defectuosos, seleccionados - 1)
    c = math.comb(defectuosos, 2) * math.comb(total - defectuosos, seleccionados - 2)
    return a, b, c

a, b, c = seleccionar_focos_defectuosos(12, 2, 3)

print("26a. Formas de seleccionar 3 focos sin defectuosos:", a)
print("26b. Formas de seleccionar 3 focos con uno defectuoso:", b)
print("26c. Formas de seleccionar 3 focos con dos defectuosos:", c)
```

# Ejercicio 32

```python
def programar_comerciales(total, repetidos=0):
    if repetidos == 0:
        return math.factorial(total)
    elif repetidos == 2:
        return math.factorial(total) // math.factorial(2)
    else:  # caso c (un comercial se repite 3 veces)
        return math.factorial(total) // math.factorial(3)

print("32a. Formas de programar 6 comerciales diferentes:", programar_comerciales(6))
print("32b. Formas de programar 6 comerciales con 2 iguales:", programar_comerciales(6, 2))
print("32c. Formas de programar 6 comerciales con 1 repetido 3 veces:", programar_comerciales(6, 3))
```
